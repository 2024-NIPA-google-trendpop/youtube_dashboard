```python
import MeCab
# MeCab 사전 경로 설정 (시스템에 맞게 경로 수정)
mecab = MeCab.Tagger("-d /opt/homebrew/lib/mecab/dic/mecab-ko-dic")

# MeCab을 통해 품사 태깅
result = mecab.parse("텍스트에서 명사를 추출합니다.")

# 결과를 줄 단위로 분리
lines = result.splitlines()

# 명사(NNG)를 필터링하여 추출
nouns = []
for line in lines:
    parts = line.split('\t')
    if len(parts) > 1 and 'NNG' in parts[1]:  # NNG: 일반 명사
        nouns.append(parts[0])

print(nouns)  # ['텍스트', '명사', '추출']
```

```python
stopwords = ['ㄱ', 'ㄴ', 'ㄷ', 'ㄹ', 'ㅁ', 'ㄹ', 'ㅂ', 'ㅅ', 'ㅇ', 'ㅈ', 'ㅊ', 'ㅋ', 'ㅌ', 'ㅍ', 'ㅎ', 
             '생각', '개', '대박', '뜻', '그날', '세상', '관련', '잉', '앞', '아이브', '이번', '이상', '자체', '날', '엠', '얘기', '최고', '월드', '사실',
             '이', '것', '그', '내', '왜', '말', '해', '나', '때', '임', '우리', '계속', '좀', '지금', '그냥', '너', '함', '뭐', '애', '저']

def extract_nouns(string, stopwords):
    parsed_string = mecab.parse(string)
    lines = parsed_string.splitlines()

    nouns = []

    for line in lines:
        parts = line.split('\t')
        if len(parts) > 1 and 'NNG' in parts[1] and parts[0] not in stopwords:  # NNG: 일반 명사
            nouns.append(parts[0])

    return nouns

sample['words'] = sample['comment'].apply(lambda x: extract_nouns(x, stopwords))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# 데이터프레임에서 각 댓글의 단어를 문자열로 변환
sample['words_str'] = sample['words'].apply(lambda x: ' '.join(x))

# TF-IDF 벡터라이저 설정 및 피팅
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(sample['words_str'])

# TF-IDF 점수를 DataFrame으로 변환
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
# 각 문서에서 상위 n개의 TF-IDF 값이 높은 단어를 추출하는 함수
def get_top_n_words(tfidf_matrix, features, top_n=5):
    top_n_words_per_doc = []
    for row in tfidf_matrix.toarray():
        top_n_indices = np.argsort(row)[-top_n:]  # 상위 n개의 인덱스 추출
        top_n_words = [features[i] for i in top_n_indices if row[i] > 0]
        top_n_words_per_doc.append(top_n_words)
    return top_n_words_per_doc

# TF-IDF 벡터라이저 설정 및 피팅
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(sample['words_str'])
features = vectorizer.get_feature_names_out()

# 상위 3개의 단어를 추출하여 sample['top_words']에 저장
sample['top_words'] = get_top_n_words(tfidf_matrix, features, top_n=3)

# 상위 3개 단어가 저장된 데이터프레임 확인
sample[['comment', 'top_words']].head(30)
```

```python
# top_words에서 빈도 계산
top_words = [word for sublist in sample['top_words'] for word in sublist]
top_words_counter = Counter(top_words)

print("\nTop words의 빈도수:")
for word, count in top_words_counter.most_common(100):
    print(f"{word}: {count}")
```

```python
from konlpy.tag import Okt, Kkma, Komoran, Hannanum

# konlpy 형태소 분석기 사용
parser = Okt()
#parser = Hannanum() - 경고
#parser = Kkma() - 에러남
#parser = Komoran() - 에러남

# 각 행의 'comment' 열의 문자열 길이를 계산한 후 100보다 큰 값들만 필터링
sample = sample[sample['comment'].str.len() > 100]

# 명사 추출
def extract_nouns(string, stopwords):
    nouns = parser.nouns(string)
    # stopwords를 제거한 명사 리스트 반환
    filtered_nouns = [noun for noun in nouns if noun not in stopwords]
    return filtered_nouns

# 예시
comment = "텍스트에서 명사를 추출합니다."
stopwords = ['추출', '자체']  # 불용어 예시
nouns = extract_nouns(comment, stopwords)
print(nouns)  # ['텍스트', '명사']
```
